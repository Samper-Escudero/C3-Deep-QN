```python
# GridBoard.py and Gridworld in the main repository have been directly copied from 
# Alexander Zai. “Deep Reinforcement Learning in Action MEAP V06”
# whereas the text in this notebook is my own implementation of the code in the book
# licenses are inherited from the ones used by A.Zai et al in the mentioned book
import numpy as np
import torch
import torch.nn as nn
from torch.nn import functional as F
from collections import OrderedDict
from torch.autograd import Variable
from Gridworld import *
from IPython.display import clear_output
import random
from matplotlib import pylab as plt
```

```python

class QNet():
    def __init__(self, n_in, n_out, n_hidden1, n_hidden2, action_set, env, gridSize = 4, gamma = 0.9, epsilon = 1.0):
        self.n_in = n_in
        self.n_out = n_out
        self.action_set = action_set
        self.gridSize = gridSize
        self.env = env
        self.state = self.getState()
        self.gamma = gamma
        self.epsilon = epsilon
        
       # self.one_hot_reward = np.ones(arms)

        # Neural network model definition 
        self.model = nn.Sequential(OrderedDict([
            ('fc1', nn.Linear(n_in, n_hidden1)),
            ('ReLu1', nn.ReLU(inplace = True)),
            ('fc2', nn.Linear(n_hidden1, n_hidden2)),
            ('ReLu2', nn.ReLU(inplace = True)),
            ('fc3', nn.Linear(n_hidden2, n_out)),
            #('ReLu3', nn.ReLU(inplace = True))
        ])
        )
                                  
    def updateEnv (self, env):
        self.env = env
    
    def getState(self):
        # Get state with some random noise
        state_ = self.env.board.render_np().reshape(1,64) + np.random.rand(1,64)/10.0 
        state = Variable(torch.from_numpy(state_).float())
        return state
    
    def getQReward(self):
        cur_reward = self.env.reward()
        # The reward of a Q_learning algorithm takes into account propect
        # So ,we need to get new state and the expected reward (maxQval)
        new_state_ = self.getState()
        with torch.no_grad(): # we do not want this pred to be considered in grads
            fut_qval = self.model(new_state_.reshape(1,64))
        
        # Are we in the endgame?
        if cur_reward == -1: 
            QReward = cur_reward + self.gamma*torch.max(fut_qval)
        else:
            QReward = cur_reward
        
        return QReward, cur_reward, new_state_
    
    def getBestAction(self, state, eps_choice = True):
        # Get reward prediction from the model
        qval = self.model(state)
        
        #av_softmax = softmax_fn(qval.data.numpy(), tau=2.0)  
        #print("av_softmax: ", av_softmax)

        #av_softmax /= av_softmax.sum() 
        if eps_choice:
            if (random.random() < self.epsilon):
                action_ = np.random.randint(0,qval.size()[1])
            else:
                action_ = (np.argmax(qval.data.numpy()))
        

        return action_, qval

    def executeAction(self, action):
        self.env.makeMove(action)
        new_state = self.getState()
        reward = self.env.reward()
        return reward, new_state
    
    def calcLoss(self, qval, update):
        y = np.zeros((1,qval.size()[1]))
        y[:] = qval.data.numpy()
        
        y[0][action_] = update
        y = Variable(torch.from_numpy(y).float())
        loss = loss_fn(qval, y)
        return loss
    
    def testReward(self):
        return self.env.reward()
    
    def setEpsilon(self, eps):
        self.epsilon = eps
        
    def displayEnv(self):
        print(self.env.display())
        
    def backpropagate(self, optimizer, loss):
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            
    def getValsFromMem(self, memory):
        old_state, action_m, reward_m, new_state_m = memory
        # Obtain qvals for each state in the batch
        old_qval = self.model(old_state) 
        newQ = self.model(new_state_m).data.numpy()
        # This time do not apply epsilon approach
        maxQ = np.max(newQ)
        # Create the expectation vector for each case
        y = torch.zeros((1,4))
        y[:] = old_qval[:]
        # calc update rule
        if reward_m == -1:
            update = (reward_m + (gamma * maxQ))
        else:
            update = reward_m
        y[0][action_m] = update
        
        return old_qval, y
```

```python
l1 = 64
l2 = 164
l3 = 150
l4 = 4

loss_fn = torch.nn.MSELoss(size_average=False)

learning_rate = 1e-4

#reinforcement learning specific parameters
gamma = 0.9 #since it may take several moves to goal, making gamma high
epsilon = 1.0

action_set = {
    0: 'u',
    1: 'd',
    2: 'l',
    3: 'r',
}
```

```python
Qnet = QNet(l1, l4, l2,l3, action_set, Gridworld(size=4, mode='static'), 4, gamma)
model = Qnet.model
```

```python
optimizer = torch.optim.Adam(Qnet.model.parameters(), lr=learning_rate)
```

```python
epochs = 1500
losses = []
batchSize = 40
buffer = 80
replay = []
max_moves = 75
h = 0

for i in range(epochs):
    Qnet.updateEnv(Gridworld(size=4, mode='static'))
    state = Qnet.getState()
    status = 1
    mov = 0
    
    #while game still in progress
    while(status == 1):
        # start movement count
        mov += 1
        # Greedily select the best action to take
        action_, qval = Qnet.getBestAction(state)
        # Execute the action, once converted to the corresponding game option
        # Retrieve the reward for such action
        reward, new_state = Qnet.executeAction(action_set[action_])

        # Experience-based training:
        
        # If we have experience enough to create a training batch, the system will be trained on a batch 
        # with the current info and past experience, randomly selected. 
        # Training is suspended until having enough experiences otherwise.
        if (len(replay) < buffer): #if buffer not filled, add to it
            replay.append((state, action_, reward, new_state))
        else: # if buffer full, overwrite old values
            replay.pop(0)
            # self.replay.append((state, action, reward, new_state_)) #we should at this after random selection
            # randomly sample our experience experiences memory
            minibatch = random.sample(replay, batchSize-1) 
            # Create X_train and Y_train variables for backward propagation
            # Add the current experience
            replay.append((state, action_, reward, new_state)) 
            minibatch.append((state, action_, reward, new_state)) 
            # Create training variables
            X_train = Variable(torch.empty(batchSize, 4, dtype=torch.float))
            y_train = Variable(torch.empty(batchSize, 4, dtype=torch.float))
            h = 0
            # for each sample in the selected batch
            for memory in minibatch:
                old_qval, y = Qnet.getValsFromMem(memory)
                # store the info in the batch
                X_train[h] = old_qval
                y_train[h] = Variable(y)
                h+=1
                
            # calc loss and Backpropagate the network
            loss = loss_fn(X_train, y_train)
            print(i, loss.item())
            Qnet.backpropagate (optimizer, loss)
            losses.append(loss.item()) 
            
        state = new_state
            
        if reward != -1 or mov > max_moves:
            status = 0
            mov = 0
        clear_output(wait=True)
        
    if epsilon > 0.1:
        epsilon -= (1/epochs)
```

```python
plt.figure(figsize=(15,10))
plt.ylabel("Loss")
plt.xlabel("Training Steps")
plt.plot(losses)
plt.savefig("dqn_1_loss_plot.pdf", format="pdf")
```

```python
def testModel(model, mode='static', display=True):
    
    epsilon = 0.05
    max_moves = 50
    win = False
    i = 0
    
    Qnet.updateEnv(Gridworld(size=4, mode=mode))
    Qnet.setEpsilon(0.05)
    state = Qnet.getState()
    status = 1
    if display: 
        print("Initial State:")
        Qnet.displayEnv()
    
    #while game still in progress
    while(status == 1):
        # Greedily select the best action to take
        action_, qval = Qnet.getBestAction(state)

        
        action = action_set[action_]
        if display: print('Move #: %s; Taking action: %s' % (i, action))
        # Execute the action, once converted to the corresponding game option
        Qnet.executeAction(action)
        state = Qnet.getState()
        if display: print(Qnet.displayEnv())
        reward = Qnet.testReward()
        if reward == 10:
            status = 0
            win = True
            if display: print("You won! Reward: {}".format(reward,))
        elif reward == -10:
            status = 0
            if display: print("Game lost; stepped into the pit. Penalty: {}".format(reward,))
        i += 1 #If we're taking more than 10 actions, just stop, we probably can't win this game
        if (i > max_moves):
            if display: print("Game lost; too many moves.")
            break
    return win
```
