```python
# GridBoard.py and Gridworld in the main repository have been directly copied from 
# Alexander Zai. “Deep Reinforcement Learning in Action MEAP V06”
# whereas the text in this notebook is my own implementation of the code in the book
# licenses are inherited from the ones used by A.Zai et al in the mentioned book
import numpy as np
import torch
import torch.nn as nn
from torch.nn import functional as F
from collections import OrderedDict
from torch.autograd import Variable
from Gridworld import *
from IPython.display import clear_output
import random
from matplotlib import pylab as plt
```

```python

class QNet():
    def __init__(self, n_in, n_out, n_hidden1, n_hidden2, action_set, env, gridSize = 4, gamma = 0.9, epsilon = 1.0):
        self.n_in = n_in
        self.n_out = n_out
        self.action_set = action_set
        self.gridSize = gridSize
        self.env = env
        self.state = self.getState()
        self.gamma = gamma
        self.epsilon = epsilon
       # self.one_hot_reward = np.ones(arms)

        # Neural network model definition 
        self.model = nn.Sequential(OrderedDict([
            ('fc1', nn.Linear(n_in, n_hidden1)),
            ('ReLu1', nn.ReLU(inplace = True)),
            ('fc2', nn.Linear(n_hidden1, n_hidden2)),
            ('ReLu2', nn.ReLU(inplace = True)),
            ('fc3', nn.Linear(n_hidden2, n_out)),
            #('ReLu3', nn.ReLU(inplace = True))
        ])
        )
                                  
    def updateEnv (self, env):
        self.env = env
    
    def getState(self):
        # Get state with some random noise
        state_ = self.env.board.render_np().reshape(1,64) + np.random.rand(1,64)/10.0 
        state = Variable(torch.from_numpy(state_).float())
        return state
    
    def getQReward(self):
        cur_reward = self.env.reward()
        # The reward of a Q_learning algorithm takes into account propect
        # So ,we need to get new state and the expected reward (maxQval)
        new_state_ = self.getState()
        with torch.no_grad(): # we do not want this pred to be considered in grads
            fut_qval = self.model(new_state_.reshape(1,64))
        
        # Are we in the endgame?
        if cur_reward == -1: 
            QReward = cur_reward + self.gamma*torch.max(fut_qval)
        else:
            QReward = cur_reward
        
        return QReward, cur_reward, new_state_
    
    def getBestAction(self, state, eps_choice = True):
        # Get reward prediction from the model
        qval = self.model(state)
        
        #av_softmax = softmax_fn(qval.data.numpy(), tau=2.0)  
        #print("av_softmax: ", av_softmax)

        #av_softmax /= av_softmax.sum() 
        if eps_choice:
            if (random.random() < self.epsilon):
                action_ = np.random.randint(0,qval.size()[1])
            else:
                action_ = (np.argmax(qval.data.numpy()))
        

        return action_, qval

    def executeAction(self, action):
        self.env.makeMove(action)
        return
    
    def calcLoss(self, qval, update):
        y = np.zeros((1,qval.size()[1]))
        y[:] = qval.data.numpy()
        
        y[0][action_] = update
        y = Variable(torch.from_numpy(y).float())
        loss = loss_fn(qval, y)
        return loss
    
    def testReward(self):
        return self.env.reward()
    
    def setEpsilon(self, eps):
        self.epsilon = eps
    def displayEnv(self):
        print(self.env.display())
```

```python
l1 = 64
l2 = 164
l3 = 150
l4 = 4

loss_fn = torch.nn.MSELoss(size_average=False)

learning_rate = 1e-4

#reinforcement learning specific parameters
gamma = 0.9 #since it may take several moves to goal, making gamma high
epsilon = 1.0

action_set = {
    0: 'u',
    1: 'd',
    2: 'l',
    3: 'r',
}
```

```python
Qnet = QNet(l1, l4, l2,l3, action_set, Gridworld(size=4, mode='static'), 4, gamma)
model = Qnet.model
```

```python
optimizer = torch.optim.Adam(Qnet.model.parameters(), lr=learning_rate)
```

```python
epochs = 1000
losses = []
for i in range(epochs):
    Qnet.updateEnv(Gridworld(size=4, mode='static'))
    #state_ = Qnet.env.board.render_np().reshape(1,64) + np.random.rand(1,64)/10.0
    state = Qnet.getState()
    status = 1
    #while game still in progress
    while(status == 1):
        
        # Greedily select the best action to take
        action_, qval = Qnet.getBestAction(state)
        # Execute the action, once converted to the corresponding game option
        Qnet.executeAction(action_set[action_])
        # Calc our reward taking into acc future income
        update, reward, new_state = Qnet.getQReward()
        # Calc the loss of the forward pass
        loss = Qnet.calcLoss(qval, update)
        # Backward computation
        print(i, loss.item())
        optimizer.zero_grad()
        loss.backward()
        losses.append(loss.item())
        optimizer.step()
        state = new_state
        # Is is the end-game?
        if reward != -1:
            status = 0
        #clear_output(wait=True)
    # Decrease epsilon as train goes, increasing exploitation and decresing exploration
    if epsilon > 0.1:
        epsilon -= (1/epochs)
```

```python
plt.figure(figsize=(15,10))
plt.ylabel("Loss")
plt.xlabel("Training Steps")
plt.plot(losses)
plt.savefig("dqn_1_loss_plot.pdf", format="pdf")
```

```python
def testModel(model, mode='static', display=True):
    
    epsilon = 0.05
    max_moves = 50
    win = False
    i = 0
    
    Qnet.updateEnv(Gridworld(size=4, mode=mode))
    Qnet.setEpsilon(0.05)
    state = Qnet.getState()
    status = 1
    if display: 
        print("Initial State:")
        Qnet.displayEnv()
    
    #while game still in progress
    while(status == 1):
        # Greedily select the best action to take
        action_, qval = Qnet.getBestAction(state)

        
        action = action_set[action_]
        if display: print('Move #: %s; Taking action: %s' % (i, action))
        # Execute the action, once converted to the corresponding game option
        Qnet.executeAction(action)
        state = Qnet.getState()
        if display: print(Qnet.displayEnv())
        reward = Qnet.testReward()
        if reward == 10:
            status = 0
            win = True
            if display: print("You won! Reward: {}".format(reward,))
        elif reward == -10:
            status = 0
            if display: print("Game lost; stepped into the pit. Penalty: {}".format(reward,))
        i += 1 #If we're taking more than 10 actions, just stop, we probably can't win this game
        if (i > max_moves):
            if display: print("Game lost; too many moves.")
            break
    return win
```
